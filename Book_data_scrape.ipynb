```python
import pandas as pd
import numpy as np
import requests 
from bs4 import BeautifulSoup
from tqdm import tqdm


lin = 'https://books.toscrape.com/catalogue/page-'
#print(lin)

Book = []
Book_details =[]

for page in tqdm(range(1,51)):
    link = lin+str(page)+'.html'
    
    res  = requests.get(link)
    soup = BeautifulSoup(res.text,'html.parser')
    for sp in soup.find_all('article',class_='product_pod'):
        image    = 'https://books.toscrape.com'+sp.find('img').get('src')[2:]
        title    = sp.find('h3').find('a').get('title')
        Book_link= 'https://books.toscrape.com/catalogue/'+sp.find('h3').find('a').get('href')
        Rating   = sp.find('p').get('class')[1]
        price    = sp.find('p',class_='price_color').text[1:]
        stock    = sp.find('p',class_='instock availability').text.strip()
        
        
        #print(av_ty)
        #print(price)
        #print(Rating)
        #print(Book_link)
        #print(title)
        #print(image)
        res           = requests.get(Book_link)
        soup          = BeautifulSoup(res.text,'html.parser')
        quantity      = soup.find('p',class_='instock availability').text.strip()[10:-1]
        #print(stock_details)
        Book.append([title,image,Book_link,Rating,price,stock,quantity])
        table = soup.find('table',class_='table table-striped').find_all('td')
        UPC            = table[0].text
        Product_Type   = table[1].text
        Price_excl_tax = table[2].text[1:]
        Price_incl_tax = table[3].text[1:]
        Tax            = table[4].text[1:]
        Reviews        = table[6].text
        
        #print(UPC)
        #print(Product_Type)
        #print(Price_excl_tax)
        #print(Price_incl_tax)
        #print(Tax)
        #print(table)
        #print(Reviews)
        Book_details.append([UPC,Product_Type,Price_excl_tax,Price_incl_tax,Tax,Reviews])
        
        
    #print(link)
```

```python
df2 = pd.DataFrame(Book_details,columns=['UPC','Product_Type','Price_excl_tax','Price_incl_tax','Tax','Reviews'])
```
